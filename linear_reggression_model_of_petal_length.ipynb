```python
import pickle
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
iris= datasets.load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df
```

```python
# sepal length, sepal width and petal width
X = df.iloc[:, [0,1,3]]
# petal length
Y = df.iloc[:, [2]]
# spliting the data
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
# building model
regressor = LinearRegression()
regressor.fit(xtrain, ytrain)
```

```python
print('The intercept is {} and the coef are {}'.format(regressor.intercept_, regressor.coef_))
```

```python
ypred = regressor.predict(xtest).tolist()
ypred_ = [val for sublist in ypred for val in sublist]


ytest['predicted'] = ypred_
ytest
```

```python
# pack the model with pickle
with open('model.pkl', 'wb') as files:
    pickle.dump(regressor, files)
```
